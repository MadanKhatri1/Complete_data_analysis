```python
import pandas as pd
import re

# Function to clean startup names
def clean_startup_name(name):
    if name.startswith('http://') or name.startswith('https://'):
        return None  
    
    name = name.strip('"')
    name = re.sub(r'\\xe2\\x80\\x99', "'", name)
    name = re.sub(r'\\+', '', name)
    name = re.sub(r'[^\w\s\'-]', '', name)
    name = ' '.join(name.split())
    
    return name


df['Startup Name'] = df['Startup Name'].apply(clean_startup_name)
df['Startup Names'] = df['Startup Names'].apply(clean_startup_name)
cleaned_names = pd.concat([df['Startup Name'], df['Startup Names']]).unique()
cleaned_names = [name for name in cleaned_names if name is not None]

sorted_cleaned_names = sorted(cleaned_names)
```

```python
unique_items_investor = []
for column in ['Investors']:
    unique_items_investor.extend([str(x) for x in df[column].dropna().unique().tolist() if x is not None])

# Sort the list
sorted_unique_investors = sorted(unique_items_investor)

# Clean up each string: replace "\\\'" and "\xc2\xa0" with a space, and optionally handle other patterns
cleaned_investors = [investor.replace("\\\'", "").replace("\xc2\xa0", " ").strip() for investor in sorted_unique_investors]

# Print the cleaned list
print(cleaned_investors)
```

```python
import re
import unicodedata

# Step 1: Get unique investor entries
unique_items_investor = []
for column in ['Investors']:
    unique_items_investor.extend([str(x) for x in df[column].dropna().unique().tolist() if x is not None])

# Step 2: Normalize Unicode characters (e.g., convert \xc3\x98 to Ø)
normalized_investors = [
    unicodedata.normalize('NFKD', investor).encode('ascii', 'ignore').decode('ascii')
    for investor in unique_items_investor
]

# Step 3: Split entries by commas and flatten the list
split_investors = [investor.split(",") for investor in normalized_investors]
flattened_investors = [item.strip() for sublist in split_investors for item in sublist if item.strip()]

# Step 4: Remove duplicates and sort
sorted_unique_investors = sorted(set(flattened_investors))

# Step 5: Clean up: remove unwanted characters, generic terms, and artifacts
cleaned_investors = [
    re.sub(
        r'[\s\\]+[\'"]?|[\'"]$|^[\'"]|[\xc2\xa0]+|xc[0-9a-f]{1,2}|xa[0-9a-f]{1,2}|[\(\)]|\.|& Others|&otherAngelinvestors|&[\w\s]*?(?:Others|Angel\s*investors|Undisclosed\s*investors|undisclosed.*?$)|others\.',
        '',
        investor,
        flags=re.IGNORECASE
    ).strip() for investor in sorted_unique_investors
]

# Step 6: Filter out empty strings, generic terms, and sort again
cleaned_investors = sorted([
    investor for investor in cleaned_investors 
    if investor and not re.match(r'^\s*$|^&.*$|^xc.*$|^unknown.*$|^undisclosed.*$', investor, re.IGNORECASE)
])

# Print the cleaned list
print(cleaned_investors)
```

```python
# Calculate the sum of amounts by vertical for IDG Ventures
import matplotlib.pyplot as plt
vertical_sums = df[df['Investors'].str.contains('IDG Ventures')].groupby("Round")['Amount'].sum().sort_values(ascending=False)

# Get the top 5 verticals
top_5 = vertical_sums.head(4)

# Sum the remaining verticals and label as "Others"
others_sum = vertical_sums.iloc[4:].sum()
if others_sum > 0:  # Only include "Others" if there are remaining verticals with non-zero sums
    top_5_with_others = pd.concat([
        top_5,
        pd.Series(others_sum, index=['Others'])
    ])
else:
    top_5_with_others = top_5

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(top_5_with_others, labels=top_5_with_others.index, autopct='%1.1f%%', startangle=90)
ax.set_title('Investment Distribution by Vertical for IDG Ventures')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
# st.pyplot(fig)
```
